```python
def quitar_retweets_y_usuarios(phrase):
    """
    :param phrase:
    :return:
    """
    caracteres_especiales = '‚Äô"?¬°!*,.():;-_¬ø'
    
    phrase = phrase.lower()
    
    phrase = re.sub("√°", "a", phrase)
    phrase = re.sub("√©", "e", phrase)
    phrase = re.sub("√≠", "i", phrase)
    phrase = re.sub("√≥", "o", phrase)
    phrase = re.sub("√∫", "u", phrase)
    
    phrase = re.sub("‚Ä¶", "", phrase)
   
    
    phrase = re.sub("rt @[\S]+: ", "", phrase)
    phrase = re.sub(" rt ", "", phrase)
    phrase = re.sub("@[\S]+", "", phrase)
    phrase = re.sub("((www\.[\S]+)|(https?://[\S]+))", "", phrase)
    regex = re.compile('[%s]' % re.escape(caracteres_especiales))
    phrase = regex.sub('', phrase)
    phrase = re.sub(' +', ' ', phrase)
    
    return phrase
```

```python
tweets_df['texto_limpio'] = tweets_df.text.map(lambda x: quitar_retweets_y_usuarios(str(x)))
```
